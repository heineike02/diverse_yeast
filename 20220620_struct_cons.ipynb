```python
seq_align_filt = AlignIO.read(fname_seq_aln_out_filt, 'fasta')

with open(fname_seq_aln_out, 'w') as f_out: 
    
    #Obtain indices of sequences that are not all gaps
    aln_length = seq_align_filt.get_alignment_length()
    N_seqs = len(seq_align_filt)
    all_gaps = '-'*N_seqs

    degap_inds = []
    for jj in range(0,aln_length):
        if not(seq_align_filt[:,jj] == all_gaps):
            degap_inds.append(jj)


    for record in seq_align_filt:
        f_out.write('>' + record.id + '\n')
        degapped_seq = ''.join([record.seq[jj] for jj in degap_inds])
        f_out.write(degapped_seq + '\n')
    
    
```

```python
#Make fasta alignment file out of amino acid mapping for CDC19



fname_struct_aln = base_dir + os.sep + os.path.normpath('alphafold\msas\\' + og + '_struct.fasta')

seqs_to_remove_struct = {'OG1299': ['alloascoidea_hylecoeti__OG1299__0_2015', #Need to remove this because of the a. hylecoeti bug
                                    'Spom_AF-Q10208-F1-model_v2'    # The sequence Alignment doesn't contain the S.pombe file.  
                                   ],
                         'OG1273': [ 'alloascoidea_hylecoeti__OG1273__0_850', 
                                     'Spom_AF-Q10062-F1-model_v2'    #https://www.uniprot.org/uniprotkb/Q10062/entry SPAC1F5.07c Pombe:  https://www.pombase.org/gene/SPAC1F5.07c
                                   ]
                        }
                             
                         
#Load aa_table
aa_table = pd.read_table(divyeast_dir + os.sep + 'aa_table.txt')
aa_dict = dict(zip(aa_table['Three_letter_caps'], aa_table['Single_letter']))

struct_aln_lines = {}
for jj, name in enumerate(og_data['Molecules'].keys()):
    if not(name in seqs_to_remove_struct[og]):
        msa_line = ''
        for aa in og_data['Mapping']['Amino Acid'][jj]: 
            aa_sl = aa_dict[aa]
            msa_line = msa_line + aa_sl
        struct_aln_lines[name] = msa_line
            
seq_align_out = AlignIO.read(fname_seq_aln_out, 'fasta')      
names_seq = [record.id for record in seq_align_out]

#checks to see if same names in both alignments
assert (len(set(names_seq)-set(struct_aln_lines.keys()))==0), 'More names in sequence based alignment than structure based alignment ' 
assert (len(set(struct_aln_lines.keys()-set(names_seq)))==0), 'More names in structure based alignment than sequence based alignment ' 

        
with open(fname_struct_aln, 'w') as f_out:
    seq_align_out = AlignIO.read(fname_seq_aln_out, 'fasta')
    for record in seq_align_out:
        name = record.id        
        f_out.write('>' + name + '\n')
        f_out.write(struct_aln_lines[name] + '\n')
```
