```python
regev_scer_growth_stress_fname = yeast_esr_exp_path + os.path.normpath('expression_data/regev_data/Scer_growth_stress_norm.csv')
regev_scer_growth_stress = pd.read_csv(regev_scer_growth_stress_fname)
regev_scer_growth_stress.rename(columns={'orf_name': 'genename'}, inplace=True)


scer_pka_inh_fname = yeast_esr_exp_path + os.path.normpath('expression_data/kl_sc_PKA_as_m24_r1g1_20181017/20181017_deseq_SC_AS_WT_nmpp1.csv')
scer_pka_inh = pd.read_csv(scer_pka_inh_fname)
scer_pka_inh.rename(columns={'Unnamed: 0': 'genename'}, inplace=True)

blevins_scer_ox_stress_fname = os.path.normpath('G:/My Drive/Crick_LMS/external_data/blevins_ox_stress/Blevins_Tavella_gene_lists_rna_prot_levels.csv')
blevins_scer_ox_stress = pd.read_csv(blevins_scer_ox_stress_fname, skiprows=1)
genename = yeast_esr_exp.SC_orf_lookup_by_name(blevins_scer_ox_stress['ID'])
blevins_scer_ox_stress['genename']=genename

reps = ['rep1','rep2','rep3']

for cond in ['normal','stress']: 
    blevins_scer_ox_stress['Proteomics_' + cond] = blevins_scer_ox_stress.loc[:,['Proteomics_' + cond + '_' + rep for rep in reps]].mean(axis=1)

blevins_scer_ox_stress['Proteomics_fc'] = blevins_scer_ox_stress['Proteomics_stress']-blevins_scer_ox_stress['Proteomics_normal']

reps = ['rep1','rep2']

for readout in ['RP', 'RNA']:
    for cond in ['normal','stress']: 
        blevins_scer_ox_stress[readout + '_' + cond] = blevins_scer_ox_stress.loc[:,[readout + '_' + cond + '_' + rep for rep in reps]].mean(axis=1)

    blevins_scer_ox_stress[readout + '_fc'] = np.log2(blevins_scer_ox_stress[readout + '_stress']/blevins_scer_ox_stress[readout + '_normal'])
```
